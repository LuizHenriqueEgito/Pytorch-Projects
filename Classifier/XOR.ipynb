```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
```

```python
from tqdm import tqdm
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
from utils import plot_decision_boundary, accuracy_fn
```

```python
# create data
X = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)
```

```python
data = torch.cat((X, y), 1)
```

```python
cmap_ = ListedColormap(['#ee3737', '#198bd9'])
```

```python
# plotting the data
fig, ax = plt.subplots()
ax.scatter(x=data[:, 0], y=data[:, 1], c=data[:, 2], s=250,cmap=cmap_, edgecolors='k')
ax.set_title('XOR plot')
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.2, 1.2)
ax.grid(True, linestyle='dotted')
plt.show()
```

```python
xor_model = nn.Sequential(
    nn.Linear(in_features=2, out_features=4),
    nn.ReLU(),
    nn.Linear(4, 1),
    nn.Sigmoid()
)
```

```python
xor_model
```

```python
# make a simple model
class XORModel(nn.Module):
    def __init__(self):
        super(XORModel, self).__init__()
        self.layer_in = nn.Linear(in_features=2, out_features=4)
        self.layer_hidden = nn.Linear(4, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.layer_in(x))
        x = self.sigmoid(self.layer_hidden(x))
        return x

# Instanciando o modelo
xor_model = XORModel()
```

```python
xor_model
```

```python
# prep the dataset using TensorDataset
dataset = TensorDataset(X, y)
# Prep DataLoader
dataloader = DataLoader(dataset, batch_size=4, shuffle=False)
```

```python
# optmize model, i will put the lr very low for see the model's training
optimizer = optim.SGD(xor_model.parameters(), lr=0.5)
```

```python
EPOCHS = 100
for epoch in tqdm(range(EPOCHS)):
    xor_model.train()
    running_loss = 0
    running_accuracy = 0
    for x_train, y_train in dataloader:
        optimizer.zero_grad()
        outputs = xor_model(x_train)
        
        loss = F.binary_cross_entropy(outputs, y_train)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        
        predicted_labels = (outputs >= 0.5).float()
        accuracy = (predicted_labels == y_train).float().mean().item()
        running_accuracy += accuracy
    
    if epoch %5 == 0:
        plot_decision_boundary(model=xor_model, s=500, color_map=cmap_ ,dataset=dataset, cols=[0, 1])
        print(f'poca: {epoch+1} | Loss: {running_loss / len(dataloader)} | acc: {running_accuracy / len(dataloader)}')
```
